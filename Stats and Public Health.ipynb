### 1.4 Are there any columns that contain duplicate information? If so, remove the redundant columns.

The `Year` and `Week` are redundant columns because they can derive from `Date` column.
So, remove both column by `df.drop` function

```python
# Drop Year and Week column
mosquitoes_df.reset_index(inplace=True)
mosquitoes_df.drop(['Year','Week'], axis=1, inplace=True)
mosquitoes_df.head()
```

```python
# Check duplicate records, there is no duplicated record
mosquitoes_df.T.duplicated()
```

### 1.5 Are there any null values in the dataframe? If so, deal with them appropriately.
**Step** 
1. Check null value percentate using `isna().mean()*100`

Only `Lat`, `Lon` columns are 15.8% null

2. Check the address of the Lat and Lon are null found 10 unique.

3. In this case, as the `Lat` and `Lon` look near each others so replace with the mode data for all missing `Lat`and `Lon`

or Try to find the actual latitude and longitude and replace with correct value. 
By this website [here](https://www.latlong.net/convert-address-to-lat-long.html) , put the address and get latitude and longitude to replace.
    

```python
# check null value percentage
mosquitoes_df.isna().mean()*100
```

```python
mosquitoes_df.loc[mosquitoes_df['Lat'].isnull()]['Address Block'].unique()
# 10 unique address has null Lat and Lon
```

```python
mosquitoes_df.loc[mosquitoes_df['Lat'].isnull()]['Mosquito number'].sum()
```

```python
# Check percentage of data and found there is 20% so cannot remove
mosquitoes_df.loc[mosquitoes_df['Lat'].isnull()]['Mosquito number'].sum() / mosquitoes_df['Mosquito number'].sum() 
```

```python
# replace by mode value

mosquitoes_df['Lat'].fillna(mosquitoes_df['Lat'].mode().values[0],inplace=True)
mosquitoes_df['Lon'].fillna(mosquitoes_df['Lon'].mode().values[0],inplace=True)
```

```python
# Check head of dataframe
mosquitoes_df.head()
```

```python
# check null value. No more!
mosquitoes_df.isna().mean()*100
```

### 2.1 Using an appropriate visual, or visuals, explore the relationship between mosquito number and date.

>Explore Year and Month which number of mosquitoes high or low by using line graph to compare each Year and Month

**Finding**
August was high number in almost every years follwing by July. August 2007 was highest. And 2016 was quite highest in every month. And no data during Nov to Jan. It might because there are a lot of mosquitoes during the summer and no mosquitoes during winter.

```python
# plot line graph shows number of mosquitoes separated by Year,Month
plt.figure()
df = mosquitoes_df.groupby([mosquitoes_df['Date'].dt.year, 
                       mosquitoes_df['Date'].dt.month])['Mosquito number'].sum().unstack().plot(kind='line', 
                                                                                                marker='.',
                                                                                                ylabel='Number of mosquitoes',
                                                                                                xlabel='Year',
                                                                                                title='Mosquitoes Over the Observed Time Period',
                                                                                                stacked=False)

plt.legend(title='Month')
plt.show()
```

**Check the time of high mosquito number.**



```python
# plot bar chart show mosquitoes number found during each hour
plt.figure()
mosquitoes_df.groupby(mosquitoes_df['Date'].dt.time)['Mosquito number'].sum().plot(kind='bar',
                                                                                   ylabel='Mosquito Number',
                                                                                   xlabel='Time',
                                                                                   title = 'Mosquito number by time')
plt.show()
```

> Found that the data contain during morning time (12 AM - 12 PM).
> The highest was 8 AM follow by 7 AM, 9 AM, and 6 AM  respectively.
> This can assume that during 6 AM - 9 AM is tend to high volumn of mosquito.

### 3.1 Using an appropriate visual, explore the relationship between mosquito species and WNV prevalence.

**Step**
1. Present as a bar chart show species and number of mosquitoes and separated WNV prevelence by stacked bar.

The most positive found 2 species "CULEX RESTUANS" and "CULEX PIPINES".

### 3.2 Using an appropriate visual, explore the relationship between the number of mosquitos caught and trap type.

Note: *This visual should be a different type of visualization than the previous one*

Below is a pie chart represent the relationship between the number of mosquitos caught and trap type.


**Check only WNV Present is Positive**
